## Criando um dataframes somente com os estados

Crio o dataframe df_states e insiro nele todos os dados do dataframe df.
A partir disso crio uma condicional onde vou excluir os campos da coluna estado que possuam o valor null e também incluo somentes os dados da coluna município que possuam valor null, fazendo assim a seleçao dos campos estado, e excluindo municípios e os dados gerais do Brasil.

```python
df_states = df[(df["estado"].notna()) & (df["codmun"].isna())]
```

```python
df_states
```

Agora iremos criar um dataframe com os dados do Brasil, sleecionando através da coluna regiao quando o valor for igual a "Brasil""


```python
df_brasil = df[df["regiao"] == "Brasil"]
```
